# General view

```python
sr_stdy = open_data("data/T_mean_1940_2025.nc", "2t")
print(sr_stdy)
```

```python
sr_test = sr_stdy.loc["2015-02-01 00:00:00":"2015-03-01 01:00:00"]
print(sr_test)
```

```python
sr_study_m = sr_stdy.resample("ME").mean()
plot_data(
    sr = sr_study_m,
    var_name = "Temperature",
    graph_title = "Monthly average temperature",
    img_title = "reanalysis/monthly_avg_temp_1940_2025",
    station=False 
)
```

```python
sr_study_y = sr_stdy.resample("YE").mean()
plot_data(
    sr = sr_study_y,
    var_name = "Temperature",
    graph_title = "Yearlyy average temperature",
    img_title = "reanalysis/yearly_avg_temp_1940_2025",
    station=False 
)
```

# Normal

```python
sr_clim_frst = sr_stdy.loc["1940-01-01 00:00:00":"1959-12-31 23:00:00"]
dayofyear = sr_clim_frst.index.dayofyear
sr_clim_mf_frst = sr_clim_frst.groupby(dayofyear).mean()

plot_data(
    sr_clim_mf_frst,
    "Normal",
    "Normal temperature reanalyzed based on 1940-1959 over Perpignan",
    "reanalysis/norm_temp_1940_1959"
)

dic_quantiles_frst = quantiles(
    sr_clim_frst,
    "Daily temperature reanalyzed quantiles based on 1940-1959 over Perpignan",
    "Temperature (C)",
    "temp/reanalysis/temp_q_1940_1959"
)
```
